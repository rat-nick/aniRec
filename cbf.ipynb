```python
from sklearn.metrics.pairwise import cosine_similarity

import pandas as pd
import config

df = pd.read_csv(config.WAREHOUSE_PATH+"/clean_items_with_features.csv", low_memory=False)
df.dtypes
```

```python
features = df.filter(regex=r"^OH").reset_index()
```

```python
ids = [7349, 9569, 5010, 18938, 5654, 19156]
sample = [features.loc[ids].sum()]
pref = df.loc[ids][["title", 'poster_url']]
pref
```

```python
sim = cosine_similarity(sample, features)
sim = list(enumerate(sim[0])) 
recIDs = sorted(sim, key= lambda x:x[1], reverse=True)
recIDs = list(map(lambda x: x[0], recIDs))
recIDs = [id for id in recIDs if id not in ids]

recs = df.loc[recIDs]

recs = recs[:50][["title", 'poster_url']]
recs
```
